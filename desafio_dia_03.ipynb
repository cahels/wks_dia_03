```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/dados (1).csv")
```

```python
df.describe() #matemática, leitura e escrita respectivamente
```

```python
#agrupamento por raça/etnia
df2 = df.groupby("race/ethnicity")[["math score"]].mean()
```

```python
df2
```

```python
df2.plot(kind = "bar")
```

```python
#agrupamento a partir da escolaridade dos pais
df3 = df.groupby("parental level of education")[["writing score"]].mean()
```

```python
df3
```

```python
df3.plot(kind = "bar")
```

```python
df4 = df.groupby("test preparation course")[["reading score"]].mean()
```
